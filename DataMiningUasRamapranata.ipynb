```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Load Dataset
Kita akan membuat dataset sederhana sebagai contoh. Dataset ini terdiri dari dua kolom, `x` dan `y`, yang merepresentasikan nilai-nilai yang akan kita klusterkan.


```python
import pandas as pd # Import the pandas library and alias it as 'pd'

data = {
    'x': [25, 34, 22, 27, 33, 46, 55, 38, 45, 50, 52, 35, 41, 43, 54, 42, 53, 48, 52, 45],
    'y': [79, 74, 55, 70, 66, 60, 55, 54, 52, 45, 40, 66, 65, 43, 42, 56, 58, 50, 46, 45]
}
df = pd.DataFrame(data) # Now you can use 'pd' to refer to the pandas library
```

## Visualisasi Dataset
Sebelum melakukan clustering, kita akan memvisualisasikan dataset untuk melihat sebaran data.


```python
plt.scatter(df['x'], df['y'])
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Dataset')
plt.show()
```

## Inisialisasi dan Fitting Model K-Means
Selanjutnya, kita akan menginisialisasi model K-Means dengan jumlah klaster sebanyak 3 dan melakukan fitting terhadap dataset.


```python
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(df)

# Menambahkan label hasil clustering ke dataframe
df['label'] = kmeans.labels_

```

## Visualisasi Hasil Clustering
Setelah model di-fit, kita akan memvisualisasikan hasil clustering. Setiap klaster akan ditampilkan dengan warna yang berbeda.


```python
colors = ['red', 'green', 'blue']
for i in range(3):
    clustered_data = df[df['label'] == i]
    plt.scatter(clustered_data['x'], clustered_data['y'], color=colors[i])

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('K-Means Clustering')
plt.show()
```
